# Human hepatic caffeine metabolism
Simple model of caffeine clearance by the human liver.

```python
# simulation to steady state
%matplotlib inline
from __future__ import print_function, division
import tellurium as te
r = te.loads('Hepatic_caffeine_1.xml')
r.integrator.absolute_tolerance = 1E-20

# print(r.getAntimony(current=False))
s = r.simulate(0,10,steps=100)
print(r.selections)
r.plot(s)
```

```python
# give a time dependent challange via composition of models
r.integrator.variable_step_size = True
for k in range(10):
    r.reset()
    r['[caf_ext]'] = k
    s = r.simulate(0, 50)
    r.plot(s, show=False, loc=False)
r.plot(s, show=False)
```
