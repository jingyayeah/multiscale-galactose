# Tolic model
Simulation of insulin glucose model

```python
%matplotlib inline
from __future__ import print_function, division
import tellurium as te
```

```python
r = te.loads("Sturis1991_1.xml")
print(r.timeCourseSelections)
```

```python
# Simulate the system
s = r.simulate(0,500, steps=500)
# r.plot(s)
```

```python
from matplotlib import pyplot as plt
```

```python
# Plasma insulin
plt.plot(s['time'], s['Ip']/(r.Vp), '-', linewidth=2, color='black')
plt.xlabel('time [min]')
plt.ylabel('Plasma insulin [mU/l]')
plt.ylim([10, 50])
plt.show()

# Plasma glucose
plt.plot(s['time'], s['G']/(r.Vg*10), '-', linewidth=2, color='black')
plt.xlabel('time [min]')
plt.ylabel('Plasma glucose [mg/dl]')
plt.ylim([100, 160])
plt.show()
```
